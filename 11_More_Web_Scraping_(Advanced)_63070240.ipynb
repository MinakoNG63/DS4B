```python
def extract_title(html_page):
  tags_title = html_page.select_one('div.product_main > h1')
  title = tags_title.text
  return title
```

```python
def extract_price(html_page):
  tags_price = html_page.select_one('div.product_main p.price_color')
  price = tags_price.text
  return price
```

```python
def extract_rating(html_page):
  tags_rating = html_page.select_one('div.product_main p.star-rating')
  rating = tags_rating["class"][1]
  return rating
```

```python
response = requests.get("http://books.toscrape.com/catalogue/soumission_998/index.html")
html_page = bs4.BeautifulSoup(response.content, 'lxml')
```

```python
def extract_image(html_page):
  tags_url = html_page.select_one('div.item.active > img')
  image_url = "http://books.toscrape.com/" + tags_url["src"][6:]
  return image_url
```

```python
def extract_desc(html_page):
  try:
    tags_desc = html_page.select_one('article.product_page > p')
    desc = tags_desc.text
    return desc
  except:
    return ''
```

```python
def extract_info(html_page):
  list_key = html_page.select('table.table.table-striped tr th')
  list_value = html_page.select('table.table.table-striped tr td')
  info_dict = dict()
  for idx in range(len(list_key)):
    info_dict[list_key[idx].text] = list_value[idx].text

  # for idx,key in enumerate(list_key):
  #   info_dict[key.text] = list_value[idx].text

  return info_dict
```

```python
response = requests.get("http://books.toscrape.com/catalogue/soumission_998/index.html")
html_page = bs4.BeautifulSoup(response.content)
```

```python
list_key = html_page.select('table.table.table-striped tr th')
list_value = html_page.select('table.table.table-striped tr td')
info_dict = dict()
for idx,key in enumerate(list_key):
    info_dict[key.text] = list_value[idx].text
    print(idx,key.text)
print(info_dict)
```

```python
#example: http://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html
def extract_book_info(url):
  response = requests.get(url)
  html_page = bs4.BeautifulSoup(response.content)

  title = extract_title(html_page)
  price = extract_price(html_page)
  rating = extract_rating(html_page)
  image_url = extract_image(html_page)
  desc = extract_desc(html_page)
  infos = extract_info(html_page)

  book_dict = {
      'title': title,
      'price': price,
      'rating': rating,
      'image_url': image_url,
      'desc': desc,
  }
  book_dict.update(infos)

  return book_dict
```

### Gather links to all books in page 1-50

```python
# extract link
links = []
for page in range(1, 51):
  # response = requests.get('http://books.toscrape.com/catalogue/page-{}.html'.format(page))
  response = requests.get(f'http://books.toscrape.com/catalogue/page-{page}.html')
  html_page = bs4.BeautifulSoup(response.content, 'lxml')

  selector = 'section article.product_pod h3 a'
  tags = html_page.select(selector)
  for i in tags:
    desc = i["href"]
    links.append(desc)
```

### Use all links to extract contents in book pages and return in json

```python
from tqdm.notebook import tqdm
```

```python
book_data = []
src = "http://books.toscrape.com/catalogue/"
for link in tqdm(links):
  print(link)
  book_dict = extract_book_info(src + link)
  book_data.append(book_dict)
```
